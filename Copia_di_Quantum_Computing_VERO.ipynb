```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit import Aer, transpile
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
from qiskit.quantum_info import random_statevector
```

#Everything is found on Quantum Computing notes of C. Sanavio on Virtuale

```python
circuit = QuantumCircuit(3,2) # circuit with 3 qubits and 2 bits
a = np.random.uniform(-1,1)+1j*np.random.uniform(-1,1)
b = np.random.uniform(-1,1)+1j*np.random.uniform(-1,1)
na = np.abs(a)
nb = np.abs(b)
a = a/np.sqrt(na**2+nb**2)
b = b/np.sqrt(na**2+nb**2)
initial_state = [a,b]
```

```python
circuit.initialize(initial_state,0)
```

```python
circuit.h(1) #Hadamard gate on qubit 1
circuit.cx(1,2) #C-X gate on qubit 2 and control qubit 1
circuit.barrier()
```

```python
circuit.draw('mpl')
```

This is the Teleportation algorithm

```python
circuit.cx(1,0)
circuit.h(1) #Hadamard gate
circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.measure(0,0)
circuit.measure(1,1)
circuit.draw('mpl')
```

```python
circuit.x(2).c_if(0, 1)
circuit.z(2).c_if(1, 1)
circuit.draw('mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
circuit.save_statevector()
out_vector = sim.run(circuit).result().get_statevector()
plot_bloch_multivector(out_vector)
```
